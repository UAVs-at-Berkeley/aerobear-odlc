The ideal input data would be a single character rotated in a cardinal direction with a monochrome background representing the shape. Initial testing with k-means on cropped targets shows that this format can be organically produced.

```python
def fitContour(img, lbound, ubound):
#     show(img)
    img_thresh = cv2.inRange(img, lbound, ubound)
#     show(img_thresh, cmap="gray")
    contours, _ = cv2.findContours(img_thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    alphanum = contours[0]
    
    vx, vy, x, y = cv2.fitLine(alphanum, cv2.DIST_L2, 0, 0.1, 0.1)
    print(vx, vy, x, y)
    
    img_line = np.array(img)
    cv2.line(img_line, (x+(img.shape[1]-1-y)/vy * vx, img.shape[1]-1), (x+(0-y)/vy * vx, 0), (255, 0, 0))
    show(img_line)
    return math.atan2(vy, vx)
```

We try a different approach: finding the rotated bounding box using cv2.minAreaRect().

```python
def fitRect(img, lbound, ubound):
    # Detect alphanum
    img_thresh = cv2.inRange(img, lbound, ubound)
    show(img_thresh, cmap="gray")
    contours, _ = cv2.findContours(img_thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    alphanum = contours[0]
    
    # Fit and display bounding box
    box = cv2.boxPoints(cv2.minAreaRect(alphanum))
    box = np.int0(box)
    cv2.drawContours(img, [box], 0, (255, 0, 0), 2)
    for i in range(4):
        pt = box[i]
        cv2.circle(img, (pt[0], pt[1]), 5, (0, 60*i, 0), -1)
    show(img)
    
    # Determine orientation using side lengths
    side1 = box[0]-box[1]
    side2 = box[-1]-box[0]
    print(np.linalg.norm(side1) - np.linalg.norm(side2))
    if abs(np.linalg.norm(side1) - np.linalg.norm(side2)) > 0.001:
        print("WOOWEE")
        if np.linalg.norm(side1) > np.linalg.norm(side2):
            return math.atan2(side1[1], side1[0])
        else:  #np.linalg.norm(side2) > np.linalg.norm(side1):
            return math.atan2(side2[1], side2[0])
    return 0
    
    
```

The fitRect method should run first in case the character is too square for fitLine to accurately determine slope. We try to round the resulting angle to some multiple of pi/4 radians, and if that doesn't work, we run fitLine as a backup:

```python
def fitCompound(img, lbound, ubound):
    # Detect alphanum
    img_thresh = cv2.inRange(img, lbound, ubound)
    show(img_thresh, cmap="gray")
    contours, _ = cv2.findContours(img_thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    alphanum = contours[0]
    
    # ATTEMPT CLASSIFICATION BY BOUNDING BOX
    img_box = np.array(img)
    box = cv2.minAreaRect(alphanum)

    # Convert output to boxpoints for drawing
    box = cv2.boxPoints(box)
    box = np.int0(box)
    cv2.drawContours(img_box, [box], 0, (255, 0, 0), 2)
    for i in range(4):
        pt = box[i]
        cv2.circle(img_box, (pt[0], pt[1]), 5, (0, 60*i, 0), -1)
    show(img_box)
    
    # Determine orientation using side lengths
    side1 = box[0]-box[1]
    side2 = box[-1]-box[0]
    print(np.linalg.norm(side1) - np.linalg.norm(side2))
    slope = 0
    # Orientation is determined by the angle of the longer side of the rectangle
    if abs(np.linalg.norm(side1) - np.linalg.norm(side2)) > 0.001:
        if np.linalg.norm(side1) > np.linalg.norm(side2):
            slope = math.atan2(side1[1], side1[0])
        else:  #np.linalg.norm(side2) > np.linalg.norm(side1):
            slope = math.atan2(side2[1], side2[0])
    else:
        # If the bounding box is sufficiently square, we have to check four
        # orientations instead of 2.
        
        # Doesn't matter which side we pick, since they're both
        # orthogonal or not orthogonal 
        slope = math.atan2(side1[1], side1[0])
        coefficient = slope / (math.pi/2)
        
        # We define orthogonality as the angle being sufficiently
        # closer to the nearest multiple of pi/2 than to the nearest
        # odd multiple of pi/4, which is denoted by the following
        # condition
        if abs(round(coefficient) - coefficient) < math.pi/8:
            print('CANDIDATES: N, W, S, E')
        else:
            print('CANDIDATES: NE, NW, SW, SE')
        return
    
    # The slope should be some multiple of pi/4; what multiple that is will tell us which orientations to look at
    coefficient = slope / (math.pi/4)
    # Threshold the accuracy of the line to the nearest multiple of pi/4.
    # If not accurate enough, we try line fitting
    if abs(round(coefficient) - coefficient) > math.pi/16:
        
        vx, vy, x, y = cv2.fitLine(alphanum, cv2.DIST_L2, 0, 0.1, 0.1)
        print(vx, vy, x, y)

        img_line = np.array(img)
        cv2.line(img_line, (x+(img.shape[1]-1-y)/vy * vx, img.shape[1]-1), (x+(0-y)/vy * vx, 0), (255, 0, 0))
        show(img_line)
        slope = math.atan2(vy, vx)
        coefficient = slope / (math.pi/4)

    coefficient = round(coefficient)
    
    if coefficient % 4 == 0:
        print('CANDIDATES: W, E')
        return "W", "E"
    elif coefficient % 4 == 1:
        print('CANDIDATES: NW, SE')
        return "NW", "SE"
    elif coefficient % 4 == 2:
        print('CANDIDATES: N, S')
        return "N", "S"
    elif coefficient % 4 == 3:
        print('CANDIDATES: NE, SW')
        return "NE", "SW"
    
        
```

```python
lbounds = [(0, 0, 200), (0, 0, 0), (0, 200, 0), (0, 0, 0)]
ubounds = [(100, 100, 255), (240, 240, 255), (255, 255, 255), (255, 100, 100)]
for i in range(4):
    img = cv2.imread('./shrubbery_orientation_test_{0}.jpg'.format(i))
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    fitCompound(img, lbounds[i], ubounds[i])
```

```python
img = cv2.imread('./shrubbery_orientation_test.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
fitCompound(img, (175, 175, 175), (255, 255, 255))
```

```python
# Function to cut out the bounding box of the alphanumeric to make sure size of template matches size of image
def cut_box(img, lbound, ubound):
    img_thresh = cv2.inRange(img, lbound, ubound)

    contours, _ = cv2.findContours(img_thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    alphanum = contours[0]
    box = cv2.boxPoints(cv2.minAreaRect(alphanum))
    box = np.int0(box)
    cv2.drawContours(img, [box], 0, (255, 0, 0), 1)
    lowx, lowy, highx, highy = box[0][1], box[0][0], box[0][1], box[0][0] 
    for y,x in box:
        if x < lowx:
            lowx = x
        if x > highx:
            highx = x
        if y < lowy:
            lowy = y
        if y > highy:
            highy = y
    return img[lowx:highx, lowy:highy]
```

```python
# File inputs (should be known before)
file_loc = './shrubbery_orientation_test_2.jpg'
# RGB values for thresholding lower and upper bound
file_lbound = (0, 200, 0) 
file_ubound = (255, 255, 255)
file_letter = '3'
orientation = ''

rec_img = cv2.imread(file_loc)
# Find the possible directions 
rec_img = cv2.cvtColor(rec_img, cv2.COLOR_BGR2RGB)

dir1, dir2 = fitCompound(rec_img, file_lbound, file_ubound)

# Trim the image 
rec_img = cut_box(rec_img, file_lbound, file_ubound)
rec_img = cv2.cvtColor(rec_img, cv2.COLOR_BGR2GRAY)

# Set up the templates
temp1_loc = './orientation-templates/' + file_letter + '-' + dir1 + '.jpg'
temp1_img = cv2.imread(temp1_loc)
temp1_img = cv2.resize(temp1_img, (int(temp1_img.shape[1] * 0.1), int(temp1_img.shape[0] * 0.1)))
temp1_img = cv2.bitwise_not(temp1_img)

temp2_loc = './orientation-templates/' + file_letter + '-' + dir2 + '.jpg'
temp2_img = cv2.imread(temp2_loc)
temp2_img = cv2.resize(temp2_img, (int(temp2_img.shape[1] * 0.1), int(temp2_img.shape[0] * 0.1)))
temp2_img = cv2.bitwise_not(temp2_img)

# Trim templates
temp1_img = cut_box(temp1_img, (175, 175, 175), (255, 255, 255))
temp1_img = cv2.cvtColor(temp1_img, cv2.COLOR_BGR2GRAY)

temp2_img = cut_box(temp2_img, (175, 175, 175), (255, 255, 255))
temp2_img = cv2.cvtColor(temp2_img, cv2.COLOR_BGR2GRAY)

# Match Templates
result1 = cv2.matchTemplate(rec_img, temp1_img, cv2.TM_CCORR)
result2 = cv2.matchTemplate(rec_img, temp2_img, cv2.TM_CCORR)

min_val1, max_val1, min_loc1, max_loc1 = cv2.minMaxLoc(result1)
min_val2, max_val2, min_loc2, max_loc2 = cv2.minMaxLoc(result2)

print(dir1, ':', max_val1)
print(dir2, ':', max_val2)

# Choose direction that has highest max_val
if max_val1 > max_val2:
    orientation = dir1
else:
    orientation = dir2
    
# Final Calculated Orientation
print('Orientation: ', orientation)
```
